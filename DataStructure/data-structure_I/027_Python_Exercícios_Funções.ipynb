1. Crie uma funcao que converta reais para dolar e euro.

```python
def converter_moeda(reais):
    # Taxas de câmbio aproximadas (exemplo: 1 USD = 5 BRL, 1 EUR = 6 BRL)
    taxa_dolar = 5.0
    taxa_euro = 6.0
    
    dolar = reais / taxa_dolar
    euro = reais / taxa_euro
    
    return dolar, euro

# Exemplo de uso
valor_em_reais = 100
dolar, euro = converter_moeda(valor_em_reais)
print(f"R$ {valor_em_reais} equivalem a ${dolar:.2f} e €{euro:.2f}")
```

2. Crie uma funcao que receba texto e retorne a quantidade de palavras.

```python
def contar_palavras(texto):
    # Divide o texto em palavras usando espaços como separador
    palavras = texto.split()
    # Retorna a quantidade de palavras
    return len(palavras)

# Exemplo de uso
texto_exemplo = "Esta é uma frase de exemplo."
quantidade = contar_palavras(texto_exemplo)
print(f"O texto '{texto_exemplo}' tem {quantidade} palavras.")
```

3. Crie uma funcao recursiva para somar de 1 ate n.

```python
def contador(n, count = 0):

	if count <= n:
		print(count)
		contador(n,count + 1)
	else:
		return count
	

contador(10,1)
		
```
