1. Compare dois produtos e informe qual tem melhor custo-beneficio (preco/quantidade).

```python
def melhor_custo_beneficio(preco1, qtd1, preco2, qtd2):
    cb1 = preco1 / qtd1
    cb2 = preco2 / qtd2
    
    if cb1 < cb2:
        return "Produto 1 tem melhor custo-benefício"
    elif cb2 < cb1:
        return "Produto 2 tem melhor custo-benefício"
    else:
        return "Os dois têm o mesmo custo-benefício"


print(melhor_custo_beneficio(10, 2, 18, 3))
```

2. Leia idade e altura e diga se a pessoa pode entrar em um brinquedo.

```python
idade = int(input("Digite a idade: "))
altura = float(input("Digite a altura (em metros): "))

if idade >= 12 and altura >= 1.40:
    print("Pode entrar no brinquedo")
else:
    print("Não pode entrar")
```

3. Verifique se um usuario pode acessar uma area restrita (idade >= 18 e ativo).

```python
idade = int(input("Digite a idade: "))
ativo = input("Usuário está ativo? (s/n): ").lower() == "s"

if idade >= 18 and ativo:
    print("Acesso permitido")
else:
    print("Acesso negado")
```
