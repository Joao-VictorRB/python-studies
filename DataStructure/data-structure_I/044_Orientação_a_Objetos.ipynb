1. Crie uma classe `Produto` com preco e metodo de desconto.

```python
class Produto:

    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def aplicar_desconto(self, percentual):
        return self.preco - (self.preco * percentual / 100)
```

2. Crie uma classe `Carrinho` que soma produtos.

```python
class Carrinho:

    def __init__(self):
        self.produtos = []

    def adicionar(self, produto):
        self.produtos.append(produto)

    def total(self):
        soma = 0
        for produto in self.produtos:
            soma += produto.preco
        return soma
```

3. Crie uma classe `Agenda` com contatos e busca por nome.

```python
class Agenda:

    def __init__(self):
        self.contatos = {}

    def adicionar(self, nome, telefone):
        self.contatos[nome] = telefone

    def buscar(self, nome):
        return self.contatos.get(nome, "Contato não encontrado")
```
