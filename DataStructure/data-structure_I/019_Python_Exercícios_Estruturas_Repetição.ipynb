1. Calcule o fatorial de um numero e mostre o passo a passo.

```python
n = int(input("Digite um número: "))

if n < 0:
    print("Fatorial não existe para números negativos.")
else:
    resultado = 1
    print(f"\nCálculo de {n}!:")

    for i in range(n, 0, -1):
        resultado *= i
        print(f"Multiplicando por {i} → parcial = {resultado}")

    print(f"\nResultado final: {resultado}")
```

2. Gere a tabuada de 1 a 10 para um numero informado.

```python
numero = int(input("Digite um número: "))

print(f"\nTabuada do {numero}:\n")

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

3. Conte quantos numeros pares e impares foram digitados (pare com 0).

```python
pares = 0
impares = 0

while True:
    numero = int(input("Digite um número (0 para sair): "))

    if numero == 0:
        break

    if numero % 2 == 0:
        pares += 1
    else:
        impares += 1

print("\nQuantidade de pares:", pares)
print("Quantidade de ímpares:", impares)
```
