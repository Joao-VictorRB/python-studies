1. Crie uma lista de compras e remova itens comprados.

```python
compras = []

while True:
    item = input("Adicionar item (ou 'sair'): ")
    if item.lower() == "sair":
        break
    compras.append(item)

print("\nLista atual:", compras)

while True:
    comprado = input("Item comprado (ou 'fim'): ")
    if comprado.lower() == "fim":
        break
    
    if comprado in compras:
        compras.remove(comprado)
        print("Removido.")
    else:
        print("Item não encontrado.")

print("\nLista final:", compras)
```

2. Monte um dicionario de contatos (nome -> telefone) e consulte por nome.

```python
contatos = {}

while True:
    nome = input("Nome (ou sair): ")
    if nome.lower() == "sair":
        break
    
    telefone = input("Telefone: ")
    contatos[nome] = telefone

print("\nConsulta de contatos")

while True:
    busca = input("Buscar nome (ou fim): ")
    if busca.lower() == "fim":
        break
    
    if busca in contatos:
        print("Telefone:", contatos[busca])
    else:
        print("Contato não encontrado.")
```

3. Encontre itens em comum entre duas listas de hobbies usando conjunto.

```python
hobbies1 = ["futebol", "leitura", "musica", "corrida"]
hobbies2 = ["musica", "filmes", "corrida", "xadrez"]

comum = set(hobbies1) & set(hobbies2)

print("Hobbies em comum:", comum)
```
