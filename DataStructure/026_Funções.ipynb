1. Crie uma funcao que calcule o frete por distancia e peso.

```python
def calcular_frete(distancia_km, peso_kg):
    if distancia_km < 0 or peso_kg < 0:
        raise ValueError("Distância e peso devem ser positivos")

    taxa_fixa = 10
    valor_por_km = 2
    valor_por_kg = 1.5

    total = taxa_fixa + (distancia_km * valor_por_km) + (peso_kg * valor_por_kg)
    return round(total, 2)
```

2. Crie uma funcao que valide uma senha (min 8 chars, 1 numero).

```python
def validar_senha(senha):
    if len(senha) < 8:
        return False
    
    tem_numero = any(char.isdigit() for char in senha)
    
    return tem_numero
```

3. Crie uma funcao que calcule estatisticas (min, max, media) de uma lista.

```python
def calcular_estatisticas(lista):
    if not lista:
        raise ValueError("Lista não pode estar vazia")

    minimo = min(lista)
    maximo = max(lista)
    media = sum(lista) / len(lista)

    return {
        "min": minimo,
        "max": maximo,
        "media": media
    }
```
