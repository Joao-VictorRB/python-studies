1. Leia notas ate o usuario digitar -1 e mostre a media.

```python
total = 0
quantidade = 0

while True:
    nota = float(input("Digite a nota (-1 para sair): "))
    
    if nota == -1:
        break
    
    total += nota
    quantidade += 1

if quantidade > 0:
    media = total / quantidade
    print(f"Média: {media:.2f}")
else:
    print("Nenhuma nota válida foi informada.")
```

2. Simule um login com 3 tentativas e bloqueio.

```python
usuario_correto = "admin"
senha_correta = "1234"

tentativas = 3

for i in range(tentativas):
    usuario = input("Usuário: ")
    senha = input("Senha: ")

    if usuario == usuario_correto and senha == senha_correta:
        print("Login realizado com sucesso!")
        break
    else:
        print(f"Credenciais inválidas. Restam {tentativas - i - 1} tentativa(s).")

else:
    print("Conta bloqueada por excesso de tentativas.")
```

3. Faça um menu simples que so termina quando o usuario escolher sair.

```python
while True:
    print("\n=== MENU ===")
    print("1 - Dizer Olá")
    print("2 - Somar dois números")
    print("0 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        print("Olá!")

    elif opcao == "2":
        a = float(input("Digite o primeiro número: "))
        b = float(input("Digite o segundo número: "))
        print("Resultado:", a + b)

    elif opcao == "0":
        print("Encerrando sistema...")
        break

    else:
        print("Opção inválida.")
```
