1. Calcule frete: se compra >= 150, frete = 0, senao frete = 20

```python
valor_compra = float(input("Digite o valor da compra: "))

if valor_compra >= 150:
    frete = 0
else:
    frete = 20

print("Frete:", frete)
```

2. Gere o status "Aprovado" ou "Recuperacao" a partir da media.

```python
media = float(input("Digite a média: "))

if media >= 7:
    status = "Aprovado"
else:
    status = "Recuperacao"

print(status)
```

3. Escolha o maior de dois numeros com operador condicional.

```python
a = float(input("Digite o primeiro número: "))
b = float(input("Digite o segundo número: "))

maior = a if a > b else b

print("Maior:", maior)
```
