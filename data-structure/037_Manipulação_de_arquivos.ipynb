1. Leia um numero com tratamento de erro para entrada invalida.

```python
try:
    numero = float(input("Digite um número: "))
    print("Número digitado:", numero)
except ValueError:
    print("Entrada inválida. Digite apenas números.")
```

2. Trate divisao por zero em uma calculadora simples.

```python
try:
    num1 = float(input("Digite o primeiro número: "))
    num2 = float(input("Digite o segundo número: "))

    resultado = num1 / num2
    print("Resultado:", resultado)

except ZeroDivisionError:
    print("Erro: não é possível dividir por zero.")
except ValueError:
    print("Erro: digite apenas números válidos.")
```

3. Tente abrir um arquivo inexistente e exiba uma mensagem amigavel.

```python
try:
    with open("arquivo.txt", "r") as arquivo:
        conteudo = arquivo.read()
        print(conteudo)

except FileNotFoundError:
    print("Arquivo não encontrado. Verifique o nome ou o caminho do arquivo.")
		
```
