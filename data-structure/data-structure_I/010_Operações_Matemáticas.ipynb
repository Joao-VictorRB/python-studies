2. Calcule consumo medio (km/l) a partir de km e litros.

```python
def consumo_medio(km, litros):
    return km / litros

print(consumo_medio(500, 40))
```

3. Calcule juros simples de um boleto atrasado (valor, taxa, dias).

```python
def calculo_juros(valor,taxa,dias):
    juros = valor * (taxa / 100) * (dias / 30)
    return juros

print(calculo_juros(1000, 5, 60))
```
