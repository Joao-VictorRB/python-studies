1. Crie um ranking de alunos (nome, nota) e ordene pela nota.

```python
alunos = [
    ("Ana", 8.5),
    ("Carlos", 7.0),
    ("Bruna", 9.2),
    ("Daniel", 6.8)
]

ranking = sorted(alunos, key=lambda x: x[1], reverse=True)

for posicao, (nome, nota) in enumerate(ranking, start=1):
    print(f"{posicao}º - {nome}: {nota}")
```

2. Conte a frequencia de palavras em um texto.

```python
texto = "python dados python engenharia dados dados"

frequencia = {}

for palavra in texto.split():
    frequencia[palavra] = frequencia.get(palavra, 0) + 1

print(frequencia)
```

3. Remova duplicados de uma lista mantendo a ordem.

```python
lista = [1, 2, 3, 2, 4, 1, 5]

vistos = set()
resultado = []

for item in lista:
    if item not in vistos:
        vistos.add(item)
        resultado.append(item)

print(resultado)
```
