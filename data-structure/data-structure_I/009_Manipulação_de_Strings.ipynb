1. Dada uma lista de nomes, padronize para "Nome Sobrenome" (primeira letra maiuscula).

```python
nomes = [
    "joao silva",
    "MARIA SOUZA",
    "  ana   costa  ",
    "carlos eduardo lima"
]

nomes_padronizados = [
    " ".join(nome.strip().split()).title()
    for nome in nomes
]

print(nomes_padronizados)
```

2. Receba um telefone e formate para (XX) XXXXX-XXXX.

```python
def formatar_telefone(telefone):
    telefone = ''.join(c for c in telefone if c.isdigit())
    return f"({telefone[:2]}) {telefone[2:7]}-{telefone[7:]}"


print(formatar_telefone("11987654321"))
```

3. Crie um "slug" de um titulo (minusculo, sem acentos, espacos viram -).

```python
def criar_slug(titulo):
    mapa_acentos = {
        "á": "a", "à": "a", "ã": "a", "â": "a",
        "é": "e", "ê": "e",
        "í": "i",
        "ó": "o", "ô": "o", "õ": "o",
        "ú": "u",
        "ç": "c"
    }
    
    titulo = titulo.lower()
    resultado = ""
    
    for char in titulo:
        if char in mapa_acentos:
            resultado += mapa_acentos[char]
        elif char.isalnum():
            resultado += char
        elif char == " ":
            resultado += "-"
    
    return resultado

print(criar_slug("Olá Mundo Python!"))
```
