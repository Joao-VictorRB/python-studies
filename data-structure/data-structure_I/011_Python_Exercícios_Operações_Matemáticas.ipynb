1. Calcule o valor de uma compra parcelada com juros mensais.

```python
def valor_parcelado(valor, juros_mensal, parcelas):
    return valor * (1 + juros_mensal) ** parcelas


print(valor_parcelado(1000, 0.02, 12))
```

2. Converta km/h para m/s e vice-versa.

```python
def kmh_para_ms(velocidade):
    return velocidade / 3.6

def ms_para_kmh(velocidade):
    return velocidade * 3.6


print(kmh_para_ms(72))   # 20.0
print(ms_para_kmh(20))   # 72.0
```

3. Leia 5 preços e mostre o menor, o maior e a media.

```python
precos = []

for i in range(5):
    valor = float(input(f"Digite o preço {i+1}: "))
    precos.append(valor)

menor = min(precos)
maior = max(precos)
media = sum(precos) / len(precos)

print("Menor:", menor)
print("Maior:", maior)
print("Média:", media)
```
