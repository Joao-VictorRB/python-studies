1. Some as vendas de uma semana e mostre total e media.

```python
total = 0

for i in range(7):
    venda = float(input(f"Venda do dia {i+1}: "))
    total += venda

media = total / 7

print(f"Total da semana: {total:.2f}")
print(f"Média diária: {media:.2f}")
```

2. Gere uma lista de boletos numerados de 1 a N.

```python
n = int(input("Quantos boletos deseja gerar? "))

for i in range(1, n + 1):
    print(f"Boleto Nº {i}")
```

3.  Conte quantas vezes aparece um valor em uma lista.

```python
valores = [10, 20, 10, 30, 10, 40]
alvo = 10

quantidade = valores.count(alvo)

print("Aparece", quantidade, "vezes")
```
