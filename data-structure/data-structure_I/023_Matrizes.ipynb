1. Simule um mapa de assentos 5x5 e marque lugares ocupados.

```python
# Criando mapa 5x5 com todos lugares livres
mapa = [["L" for _ in range(5)] for _ in range(5)]

# Marcando alguns lugares como ocupados
ocupados = [(0, 1), (2, 3), (4, 4), (1, 0)]

for linha, coluna in ocupados:
    mapa[linha][coluna] = "X"

# Exibindo mapa
for linha in mapa:
    print(" ".join(linha))
```

2. Leia uma matriz 3x3 e calcule a soma das linhas.

```python
matriz = []

# Leitura da matriz
for i in range(3):
    linha = list(map(int, input(f"Digite 3 números da linha {i+1}: ").split()))
    matriz.append(linha)

# Soma das linhas
for i in range(3):
    soma = sum(matriz[i])
    print(f"Soma da linha {i+1}: {soma}")
```

3. Encontre o maior valor e sua posicao em uma matriz 4x4.

```python
matriz = []

# Leitura da matriz 4x4
for i in range(4):
    linha = list(map(int, input(f"Linha {i+1}: ").split()))
    matriz.append(linha)

# Inicializa com o primeiro elemento
maior = matriz[0][0]
posicao = (0, 0)

# Percorrendo a matriz
for i in range(4):
    for j in range(4):
        if matriz[i][j] > maior:
            maior = matriz[i][j]
            posicao = (i, j)

print(f"Maior valor: {maior}")
print(f"Posição: linha {posicao[0]}, coluna {posicao[1]}")
```
