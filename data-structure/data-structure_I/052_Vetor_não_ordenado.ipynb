1. Crie um cadastro simples de clientes em vetor nao ordenado.

2. Implemente busca linear e conte comparacoes.

3. Remova um cliente pelo id e reorganize o vetor.

```python
class Cliente:
    def __init__(self, id, nome):
        self.id = id
        self.nome = nome


class CadastroClientes:
    def __init__(self, capacidade):
        self.clientes = [None] * capacidade
        self.tamanho = 0

    def adicionar(self, cliente):
        if self.tamanho < len(self.clientes):
            self.clientes[self.tamanho] = cliente
            self.tamanho += 1
        else:
            print("Cadastro cheio!")

    def buscar(self, id):
        comparacoes = 0

        for i in range(self.tamanho):
            comparacoes += 1
            if self.clientes[i].id == id:
                print(f"Comparações: {comparacoes}")
                return self.clientes[i]

        print(f"Comparações: {comparacoes}")
        return None

    def remover(self, id):
        for i in range(self.tamanho):
            if self.clientes[i].id == id:

                for j in range(i, self.tamanho - 1):
                    self.clientes[j] = self.clientes[j + 1]

                self.clientes[self.tamanho - 1] = None
                self.tamanho -= 1
                return True

        return False

    def listar(self):
        for i in range(self.tamanho):
            c = self.clientes[i]
            print(f"{c.id} - {c.nome}")


cadastro = CadastroClientes(10)

cadastro.adicionar(Cliente(1, "Ana"))
cadastro.adicionar(Cliente(2, "Carlos"))
cadastro.adicionar(Cliente(3, "João"))

cadastro.listar()

print("\nBuscando ID 2:")
cadastro.buscar(2)

print("\nRemovendo ID 2:")
cadastro.remover(2)

cadastro.listar()
```
