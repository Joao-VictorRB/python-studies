1. Classifique o IMC em magreza, normal, sobrepeso ou obesidade.

```python
peso = float(input("Digite o peso (kg): "))
altura = float(input("Digite a altura (m): "))

imc = peso / (altura ** 2)

if imc < 18.5:
    classificacao = "Magreza"
elif imc < 25:
    classificacao = "Normal"
elif imc < 30:
    classificacao = "Sobrepeso"
else:
    classificacao = "Obesidade"

print(f"IMC: {imc:.2f}")
print("Classificação:", classificacao)
```

2. Calcule a tarifa de energia por faixas de consumo

```python
consumo = float(input("Digite o consumo (kWh): "))

if consumo <= 100:
    valor = consumo * 0.50
elif consumo <= 200:
    valor = consumo * 0.75
else:
    valor = consumo * 1.00

print(f"Total: R$ {valor:.2f}")
```
